**Univariate analysis**

Provides summary statistics on only **one** variable.

```python
# 1-D Scatter Plot
setosa = df.loc[df["species"] == "Iris-setosa"]
versicolor = df.loc[df["species"] == "Iris-versicolor"]
virginica = df.loc[df["species"] == "Iris-virginica"]

plt.plot(setosa['petal_length'], np.zeros_like(setosa['petal_length']), 'o')
plt.plot(versicolor['petal_length'], np.zeros_like(versicolor['petal_length']), 'o')
plt.plot(virginica['petal_length'], np.zeros_like(virginica['petal_length']), 'o')

plt.grid()
plt.show()
```

**blue** - Iris-setosa

**orange** - Iris-versicolor

**green** - Iris-virginica

Iris-setosa is **linearly separable**(separated by a line)

**if (petal_length <= 2) Iris-setosa**

**else Not Iris-setosa**


```python
sns.boxplot(x="species",y="petal_length",data=df)
plt.show()
```

```python
sns.violinplot(x="species",y="petal_length",data=df)
plt.show()
```

```python
sns.distplot([])
```

**Biraviate Analysis**

```python
# 2-D Scatter Plot
sns.set_style('whitegrid')
sns.FacetGrid(df,hue="species",height=4).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.show()
```

Iris-virginica and Iris-versicolor have an overlap and are not linearly separable.

```python
sns.relplot(x='sepal_length',y='sepal_width', hue='species' , data = df)
```

**Multivariate Analysis**

```python
# 3-D plot
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(df,x= 'sepal_length', y= 'sepal_width', z= 'petal_width', color= 'species')
fig.show()
```
